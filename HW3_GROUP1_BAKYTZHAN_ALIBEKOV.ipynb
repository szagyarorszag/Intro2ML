Express $\sigma''(x)$ in terms of $\sigma(x)$ where $\sigma(x) = \frac 1{1 + e^{-t}}$ â€” sigmoid function. Find all $x$ for which $\sigma''(x) = 0$. Also find intervals on which sigmoid is convex/concave.

### MY SOLUTION:

$\text{Let's find }\sigma''(x) \text{ in terms of }\sigma(x) \text{ and find roots of }\sigma''(x) = 0\text{ , where }\sigma(x) = \frac{1}{1 + e^{-x}}\text{ is the sigmoid function. }$

$\sigma'(x) = \frac{d}{dx} \left( \frac{1}{1 + e^{-x}} \right) = \frac{e^{-x}}{(1 + e^{-x})^2} = \sigma(x)(1 - \sigma(x))$

$\sigma''(x) = \frac{d}{dx} \left[ \sigma'(x) \right] = \frac{d}{dx} \left[ \sigma(x)(1 - \sigma(x)) \right]$

$\text{By using dot product rule we can get } - \sigma''(x) = \sigma'(x)(1 - \sigma(x)) + \sigma(x)(-\sigma'(x))$

$\text{By using this }\sigma'(x) = \sigma(x)(1 - \sigma(x)) \text{ We can get ->}$

$\sigma''(x) = \sigma(x)(1 - \sigma(x))^2 - \sigma(x)^2(1 - \sigma(x))$

$\text{Finally }\sigma''(x) = \sigma(x)(1 - \sigma(x))(1 - 2\sigma(x))$

$\text{So we can get 3 equations to find "Roots" of the second derivative. Which is -  } \sigma''(x) = 0$

$\text{Fisrt one is } \sigma(x)=0. \text{Which means }x\text{ is infinitively small number}$

$\text{Second one is } 1 - \sigma(x)=0. \text{Which means }\sigma(x)=1\text{ and }x\text{ is infinitively big number}$

## Task 3.2 (0.5 points)

Let $f(\boldsymbol X) = \vert \boldsymbol X \vert^k$, $k\in\mathbb N$. Find $\nabla f(\boldsymbol X)$.

### MY SOLUTION:

$\text{Since }\boldsymbol X = (x_1, x_2, ..., x_n) \text{ and } k\in\mathbb N, \text{ also } $
$\nabla f(\boldsymbol X) = \left(\frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, ..., \frac{\partial f}{\partial x_n}\right) $

$\vert \boldsymbol X \vert = \sqrt{x_1^2 + x_2^2 + ... + x_n^2}, \text{ Then } f(\boldsymbol X) = \sqrt{x_1^2 + x_2^2 + ... + x_n^2}^k$ 

$\text{So } \nabla f(\boldsymbol X) = k (\sqrt{x_1^2 + x_2^2 + ... + x_n^2})^{k-1} \cdot \frac{1}{2\sqrt{x_1^2 + x_2^2 + ... + x_n^2}} \cdot 2x_i $

$ \nabla f(\boldsymbol X) = kx_i(\sqrt{x_1^2 + x_2^2 + ... + x_n^2})^{k-2}$

$\text{Finally }\nabla f(\boldsymbol X) = k(\sqrt{x_1^2 + x_2^2 + ... + x_n^2})^{k-2} \cdot \boldsymbol X$

## Task 3.3 (2 points)

Let $\boldsymbol A \in \mathbb R^{n\times n}$ be a symmetric matrix and

$$
f(\boldsymbol x) = \frac{\boldsymbol x^\mathsf{T} \boldsymbol{Ax}}{\boldsymbol x^\mathsf{T} \boldsymbol x}, \quad \boldsymbol x \in \mathbb R^n.
$$ 

(a) (**1 point**) Calculate $\nabla f(\boldsymbol x)$ and prove that $\nabla f(\boldsymbol x) = \boldsymbol 0$ iff $\boldsymbol x$ is an eigenvector of $\boldsymbol A$.

(b) (**1 point**) Prove that if $\boldsymbol A$ is positive definite then

$$
\max\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\max}(\boldsymbol A), \quad
\min\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\min}(\boldsymbol A)
$$

### MY SOLUTION:
(a)

$\text{Let's denote }g(\boldsymbol x) = \boldsymbol x^\mathsf{T} \boldsymbol{Ax} \text{ and } h(\boldsymbol x) = \boldsymbol x^\mathsf{T} \boldsymbol x. \text{ Then } f(\boldsymbol x) = \frac{g(\boldsymbol x)}{h(\boldsymbol x)}$.     

$\text{Then }\nabla f(\boldsymbol x) = \frac{h(\boldsymbol x) \nabla g(\boldsymbol x) - g(\boldsymbol x) \nabla h(\boldsymbol x)}{[h(\boldsymbol x)]^2}$

$\text{And then if we will calculcate }\nabla g(x) \text{ and } \nabla h(x) :  $

$\nabla g(\boldsymbol x) = \nabla (\boldsymbol x^\mathsf{T} \boldsymbol{Ax}) = \boldsymbol A \boldsymbol x + \boldsymbol A^\mathsf{T} \boldsymbol x = 2\boldsymbol A \boldsymbol x$

$\nabla h(\boldsymbol x) = \nabla (\boldsymbol x^\mathsf{T} \boldsymbol x) = 2\boldsymbol x$ 

$\text{And then if we will substitute }\nabla f(\boldsymbol x) = $
$ \frac{2\boldsymbol A \boldsymbol x \boldsymbol x^\mathsf{T} \boldsymbol x - 2\boldsymbol x \boldsymbol x^\mathsf{T} \boldsymbol{Ax}}{(\boldsymbol x^\mathsf{T} \boldsymbol x)^2} = $
$\frac{2\boldsymbol A \boldsymbol x - 2f(\boldsymbol x)\boldsymbol x}{\boldsymbol x^\mathsf{T} \boldsymbol x}$

$ \text{ which means} \boldsymbol x \text{ is an eigenvector of }\boldsymbol A$.


(b)

$ \text{If } \boldsymbol A \text{ is positive, hence eigenvalues are positive, } \lambda_{\max}(\boldsymbol A) \text{ and} \lambda_{\min}(\boldsymbol A) \text{ be the max and min eigenvalues of  }\boldsymbol A$ 

$ \text{For any }\boldsymbol x \ne \boldsymbol 0, \text{ we will have }$


$\lambda_{\min}(\boldsymbol A) \le \frac{\boldsymbol x^\mathsf{T} \boldsymbol{Ax}}{\boldsymbol x^\mathsf{T} \boldsymbol x} \le \lambda_{\max}(\boldsymbol A). $


$\text{It is equal iff } \boldsymbol x \text{ is the corresponding eigenvector of }\boldsymbol A$.


$\max\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\max}(\boldsymbol A), \quad $

$\min\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\min}(\boldsymbol A).$
