## Task 2.1 (0.5 points)

Show that similar matrices have equal determinants: $\det \boldsymbol A = \det \boldsymbol B$ if $\boldsymbol A \sim \boldsymbol B$.

### MY SOLUTION HERE:



$\text{Suppose we have two matrices } \boldsymbol{A} \text{ and } \boldsymbol{B} \text{ with size }n\times{n} \text{ } \boldsymbol{A} \text{ and } \boldsymbol{B}  \in\mathbb R^{n\times n} $



$\boldsymbol{A} \text{ and } \boldsymbol{B} \text{ are similar if there exists matrix } \boldsymbol{P} \text{ such that } \boldsymbol{A} = \boldsymbol{P}^{-1}\boldsymbol{B}\boldsymbol{P}$

$\text{Then}$ 


$\det(A) = \det(P^{-1}BP) = \det(P^{-1})\det(B)\det(P)$


$\text{Since} \det(P^{-1}) = \frac{1}{\det(P)} \text{, we have:}$


$\det(A) = \frac{1}{\det(P)}\det(B)\det(P) = \det(B)$


$\text{Thus, }\det A = \det \text{ B for similar matrices A and B}
$

## Task 2.2 (0.5 points)

Prove that all eigenvalues of a projection matrix
are equal either to $0$ or $1$.

### MY SOLUTION HERE:
$\text{Let P be a projection matrix. An eigenvalue }\lambda \text{ and its corresponding eigenvector }  v \text{ satisfy } Pv =  \lambda v$

$\text{Since  }P^2 = P \text{ , we have } P(Pv) = P^2v = Pv \text{ Thus, }Pv - \lambda  \text{ }Pv = 0 \text{, which simplifies to }(1 - \lambda)Pv = 0$.

$\text{If v is a nontrivial eigenvector, then Pv cannot be zero, and hence, we must have }1 - \lambda = 0 \text{ , which implies }\lambda = 1$.

$\text{Therefore, all eigenvalues of a projection matrix are either }\lambda = 0 \text{ or } \lambda = 1$.

[Wikipedia](https://en.wikipedia.org/wiki/Hilbert_matrix) says that the condition number of $n\times n$ Hilbert matrix grows as $O\Big(\frac 1{\sqrt n}\big(1 + \sqrt{2}\big)^{4n}\Big)$. Check this fact numerically. Calculate $\kappa_n$ for $n= 1, \ldots, 20$, and plot its graph along with the graph of the sequence inside big-O.

```python
import matplotlib.pyplot as plt
import numpy as np
import mpmath 

%config InlineBackend.figure_format = 'svg'

def hilbert_condition_number(n):
    H = mpmath.hilbert(n)
    cond_number = mpmath.cond(H)
    return cond_number
def theoretical_bound(n):
    return ( (1 + np.sqrt (2))**(4 * n) ) * 1 / ( np.sqrt(n))
def plot_hilbert_cond(n_max):
    ns = np.arange(1, n_max )
    condition_numbers = [hilbert_condition_number(n) for n in ns]
    theoretical_bound_values = [theoretical_bound(n) for n in ns]
    plt.plot(ns, condition_numbers, label=r'$\kappa_n$')
    plt.plot(ns, theoretical_bound_values, label=r'$0\left(\frac{1}{\sqrt{n}}(1+\sqrt{2})^{4n}\right)$', linestyle='dashed')
    plt.xticks(np.arange(1, n_max + 1, step=1))
    plt.xlim(0, n_max)
    plt.title("Hilbert condition number")
    plt.legend()
    plt.xlabel("n")
    plt.ylabel("Condition Number (log scale)")
    plt.yscale('log' )
    plt.grid(ls=":")
plot_hilbert_cond(20)



```

## Task 2.4 (1 point)

Let $\boldsymbol u_1, \ldots, \boldsymbol u_r \in\mathbb R^{n}$ be an orthonormal system of vectors, $1 \leqslant r \leqslant n$, and

$$
\boldsymbol A = \sum\limits_{i=1}^r \sigma_i \boldsymbol u_i \boldsymbol u_i^{\mathsf T}, \quad \sigma_1 \geqslant \sigma_2 \geqslant \ldots \geqslant \sigma_r > 0.
$$

Show that $\boldsymbol A$ is symmetric and semi-positive definite, i.e., $\boldsymbol A^\mathsf{T} =\boldsymbol A$ and $\boldsymbol x^\mathsf{T}\boldsymbol{Ax} \geqslant 0$ for all $\boldsymbol x \in\mathbb R^n$. Under which condition $\boldsymbol A$ is positive definite?

### MY SOLUTION  HERE:
$ \text{To show that }\boldsymbol{A} \text{ is symmetric, we note that }\boldsymbol{A} = \sum_{i=1}^r \sigma_i \boldsymbol{u}_i \boldsymbol{u}_i^\mathsf{T} \text{ and observe that taking the transpose }$
$\text{gives }\boldsymbol{A}^\mathsf{T} = \sum_{i=1}^r \sigma_i (\boldsymbol{u}_i \boldsymbol{u}_i^\mathsf{T})^\mathsf{T} = \sum_{i=1}^r \sigma_i \boldsymbol{u}_i \boldsymbol{u}_i^\mathsf{T} = \boldsymbol{A} \text{, showing symmetry}.$

$\text{To demonstrate semi-positive definiteness, for any } \boldsymbol{x} \in \mathbb{R}^n \text{, we have } \boldsymbol{x}^\mathsf{T}\boldsymbol{A}\boldsymbol{x} = \sum_{i=1}^r \sigma_i (\boldsymbol{x}^\mathsf{T}\boldsymbol{u}_i)^2 \geqslant 0$ since $\sigma_i > 0$ and $(\boldsymbol{x}^\mathsf{T}\boldsymbol{u}_i)^2 \geqslant 0$.

$\boldsymbol{A} \text{ is positive definite if all }\sigma_i > 0 \text{ , ensuring }\boldsymbol{x}^\mathsf{T}\boldsymbol{A}\boldsymbol{x} > 0 \text{ for all non-zero }\boldsymbol{x} \in \mathbb{R}^n$.

## Task 2.5 (1 point)

Take a **colored** image which has at least $300$ pixels in both width and height and apply SVD like it is done {ref}`here <image-svd>`. Display several low-rank approximations to your image and plot the graph of errors $\Vert \boldsymbol A - \hat{\boldsymbol A}_r\Vert_F$ for all possible $r$.

An image can be read, for example, by `plt.imread`:

```python
import matplotlib.pyplot as plt 
import numpy as np

img = plt.imread(".\la.jpg")
X = np.array(img)
plt.imshow(img)
X = np.mean(X, -1)

def compress_img(X, r):
    U, sigma, V = np.linalg.svd(X, full_matrices=True)
    x_hat = np.dot(np.dot(U[:, :r], np.diag(sigma[:r])), V[:r, :])  
    plt.imshow(x_hat, cmap='gray')
```

```python
compress_img(X,1)
```

```python
compress_img(X,4)
```

```python
compress_img(X,10)
```

```python
compress_img(X,15)
```

```python
import numpy as np

errors = []
U, S, VT = np.linalg.svd(X, full_matrices=False)

for r in range(1, min(X.shape[0], X.shape[1]) + 1): 
    reconstructed_image = U[:, :r] @ np.diag(S[:r]) @ VT[:r, :]  

    error = np.linalg.norm(X - reconstructed_image, ord='fro')  
    errors.append(error)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(errors) + 1), errors, marker='o', linestyle='-')
plt.xlabel('r (Rank)')
plt.ylabel('Frobenius Norm Error')
plt.title('Error vs. Rank for Low-Rank Approximations')
plt.grid(True)
plt.show()
```
